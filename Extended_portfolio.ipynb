```python
import csv

#initialise variables
arr = []
maxprofit=0;
minprofit=0;
avgprofit=0;
sumprofit=0;
arrindex=0;
aocindex=0;
arraocountry=[];
arraoprofit=[];
reader = csv.DictReader(open("elp_sales.csv",'rU'), dialect=csv.excel_tab, delimiter=',')
for row in reader:
    #print(row)
    arr.append(row)
    arrindex=arrindex+1;
maxprofit=float(arr[0]['Total Profit']);  #calculating max profit
minprofit=float(arr[0]['Total Profit']);  #calculating min profit
for i in range(0,arrindex):
    
    if((arr[i]['Region'].strip())=="Australia and Oceania"):
      aocindex=aocindex+1;
      sumprofit=sumprofit+float(arr[i]['Total Profit']);
      arraocountry.append(arr[i]['Country'] + "-" + (arr[i]['Item Type']));  #extract name of country and item type
      arraoprofit.append(arr[i]['Total Profit']);  #calculating profit
      if(float(arr[i]['Total Profit']) > maxprofit):
        maxprofit=float(arr[i]['Total Profit']); #maximum profit
        
      else:
        if(float(arr[i]['Total Profit']) < minprofit):
          minprofit=float(arr[i]['Total Profit']); #minimum profit
avgprofit=round((sumprofit/aocindex),2);  #calculation of average profit
print("Minimum profit: " + str(minprofit)+ " " + "Maximum profit:" + str(maxprofit)+ " " + "Average profit: " + str(avgprofit));   
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np
def graphplotter(country,totalprofit):
  colors = ['#ffc0cb', '#1f77b4', '#42c0ab','#9f7db8', '#0ca9dc','#00855b','#8a4e5d','#6a84c1','#334693','#e2e357','#653700']
 
  fig, ax = plt.subplots(figsize=(10, 10))
  # Pie Chart
  plt.pie(totalprofit, colors=['red','green','blue','skyblue'],
          autopct='%1.1f%%', pctdistance=0.85, textprops={'color':"w"})

  # draw circle
  centre_circle = plt.Circle((0, 0), 0.70, fc='white')
  fig = plt.gcf()

  # Adding Circle in Pie chart
  fig.gca().add_artist(centre_circle)

  # Adding Title of chart
  plt.title('Total profit for the region of Australia and Oceania')

  #plt.setp(autotexts, size=8, weight="bold")


  def func(pct, allvals):
      absolute = int(pct/100.*np.sum(allvals))
      return "{:.1f}%\n({:d} g)".format(pct, absolute)


  wedges, texts, autotexts = ax.pie(totalprofit, autopct=lambda pct: func(pct, totalprofit),
                                    textprops=dict(color="w"))

  ax.legend(wedges, country,
            title="Country",
            loc="center left",
            bbox_to_anchor=(1, 0, 0.5, 1))



  # Displaying Chart
  plt.show()
for rows in arraocountry:
    print(rows);
arraoprofit = [float(i) for i in arraoprofit]
graphplotter(arraocountry,arraoprofit); #plotting the graph
```

```python
import random

def askForGuess():
 guess = input(' > ')
 while(not(guess.isdecimal()) or (int(guess)<0 or int(guess)>100)):
    print('Please enter a number between 1 and 100.') #checking if guess is in the range 
    guess = input(' > ') #initialising guess to '>'
 return int(guess);

SecretNumber = random.randint(1, 100) #random package 
print('I am thinking of a number between 1 and 100.')

for i in range(0,10): 
    print('You have' + " " + str(10 - i) + " " + 'guesses left. Take a guess.') #number of guess left is displayed but only as string; thus it must be converted

    guess = int(askForGuess()) #assigning values to guess
    if guess == SecretNumber: #compares guess if it matches with the secret number
        break

    if guess > SecretNumber:  
        print('Your guess is too high.')  #guess is greater than the secret number
    if guess < SecretNumber:
        print('Your guess is too low.')  #guess is lower than secret number

if guess == SecretNumber:
    print('You guessed my number!') #output message if the secret number is rightly guessed
else:
    print('No more guesses. The number I was thinking of was', SecretNumber)
```
